```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_val0_cost100 = pd.read_csv("salidas_ponderaciones_csv/val0-cost100.csv")
df_val25_cost75 = pd.read_csv("salidas_ponderaciones_csv/val25-cost75.csv")
df_val50_cost50 = pd.read_csv("salidas_ponderaciones_csv/val50-cost50.csv")
df_val75_cost25 = pd.read_csv("salidas_ponderaciones_csv/val75-cost25.csv")
df_val100_cost0 = pd.read_csv("salidas_ponderaciones_csv/val100-cost0.csv")

df_val0_cost100.columns = ['Iteracion', 'Solucion', 'Fitness']
df_val25_cost75.columns = ['Iteracion', 'Solucion', 'Fitness']
df_val50_cost50.columns = ['Iteracion', 'Solucion', 'Fitness']
df_val75_cost25.columns = ['Iteracion', 'Solucion', 'Fitness']
df_val100_cost0.columns = ['Iteracion', 'Solucion', 'Fitness']
```

Graficos

```python
fig, axs = plt.subplots(2, 3, figsize=(15, 10))

axs[0, 0].plot(df_val0_cost100['Iteracion'], df_val0_cost100['Fitness'], marker='o', linestyle='-', color='blue', label='Fitness v/s Iterations')
axs[0, 0].set_xlabel('Iteracion')
axs[0, 0].set_ylabel('Fitness')
axs[0, 0].set_title('Fitness v/s Iteracion valor 0 - costo 100')
axs[0, 0].grid(True)

axs[0, 1].plot(df_val25_cost75['Iteracion'], df_val25_cost75['Fitness'], marker='o', linestyle='-', color='red', label='Fitness v/s Iteracion')
axs[0, 1].set_xlabel('Iteracion')
axs[0, 1].set_ylabel('Fitness')
axs[0, 1].set_title('Fitness v/s Iteracion valor 25 - costo 75')
axs[0, 1].grid(True)

axs[0, 2].plot(df_val50_cost50['Iteracion'], df_val50_cost50['Fitness'], marker='o', linestyle='-', color='green', label='Fitness v/s Iteracion')
axs[0, 2].set_xlabel('Iteracion')
axs[0, 2].set_ylabel('Fitness')
axs[0, 2].set_title('Fitness v/s Iteracion valor 50 - costo 50')
axs[0, 2].grid(True)

axs[1, 0].plot(df_val75_cost25['Iteracion'], df_val75_cost25['Fitness'], marker='o', linestyle='-', color='orange', label='Fitness v/s Iteracion')
axs[1, 0].set_xlabel('Iteracion')
axs[1, 0].set_ylabel('Fitness')
axs[1, 0].set_title('Fitness v/s Iteracion valor 75 - costo 25')
axs[1, 0].grid(True)

axs[1, 1].plot(df_val100_cost0['Iteracion'], df_val100_cost0['Fitness'], marker='o', linestyle='-', color='purple', label='Fitness v/s Iteracion')
axs[1, 1].set_xlabel('Iteracion')
axs[1, 1].set_ylabel('Fitness')
axs[1, 1].set_title('Fitness v/s Iteracion valor 100 - costo 0')
axs[1, 1].grid(True)

fig.delaxes(axs[1, 2])

plt.tight_layout()
plt.show()
```

```python
fig2, axs2 = plt.subplots(2, 3, figsize=(13, 8))

axs2[0, 0].boxplot(df_val0_cost100['Fitness'], tick_labels=['Fitness'])
axs2[0, 0].set_title('Fitness v/s Iteracion valor 0 - costo 100')
axs2[0, 0].grid(True)

axs2[0, 1].boxplot(df_val25_cost75['Fitness'], tick_labels=['Fitness'])
axs2[0, 1].set_title('Fitness v/s Iteracion valor 25 - costo 75')
axs2[0, 1].grid(True)

axs2[0, 2].boxplot(df_val50_cost50['Fitness'], tick_labels=['Fitness'])
axs2[0, 2].set_title('Fitness v/s Iteracion valor 50 - costo 50')
axs2[0, 2].grid(True)

axs2[1, 0].boxplot(df_val75_cost25['Fitness'], tick_labels=['Fitness'])
axs2[1, 0].set_title('Fitness v/s Iteracion valor 75 - costo 25')
axs2[1, 0].grid(True)

axs2[1, 1].boxplot(df_val100_cost0['Fitness'], tick_labels=['Fitness'])
axs2[1, 1].set_title('Fitness v/s Iteracion valor 100 - costo 0')
axs2[1, 1].grid(True)

fig2.delaxes(axs2[1, 2])

plt.tight_layout()
plt.show()
```
